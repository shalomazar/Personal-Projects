```python
# run forest run
%matplotlib inline

import numpy as np

# let's sample some data
n = 50
base_vec = np.linspace(0, 100, n)
noise_vec = np.random.normal(0, 3, n)

# plotting
f = plt.figure()
plt.plot(base_vec, '.g', label='base')
plt.plot(base_vec+noise_vec, 'b', label='with noise')
plt.legend()
plt.title('A simple example')
plt.show()
```

# Scattered
lines are great but lose their appeal after a while. First let's see a motivational example describing the location and quality of food on a 2D plane

```python
import pandas as pd

# generating some random data
n = 50

# coordinates
x = np.random.normal(0, 0.1, n) + np.random.randint(-1, 1, n)
y = np.random.normal(0, 0.13, n) + np.random.randint(-1, 1, n)

# food features
tasty = np.random.randint(0, 5, n) * 100
healthy = np.random.poisson(1, n) * 100

looking_for_food = pd.DataFrame({'x': x, 'y':y, 'tasty': tasty, 'healthy': healthy})
looking_for_food.head()
```

All these numbers are making me dizzy. Where's the food? Let's find out using a scatter plot. A scatter plot is a 2D plot for showing how points are scattered on a 2D plane. Run the next example to see it in action.

```python
f = plt.figure()

plt.scatter(looking_for_food.x, looking_for_food.y)
plt.show()
```

OK. This is quite informative. We now have a sense of where the food is at and how spread out is it. But... is it tasty? How can we use a 2D plot to show us? It's not hard. Let's add some color with the `c` keyword to scatter.

```python
f = plt.figure()

plt.scatter(looking_for_food.x, looking_for_food.y, c=looking_for_food.tasty, s=looking_for_food.healthy, alpha=.3)
plt.show()
plt.colorbar()
```

#### This is nice and cheery, but what do the colors mean? So much like labels and the `legend` command, colors are not very useful if we don't show a colorbar using the `plt.colorbar` function.

# Histograms
histograms are an awesome tool to explore distributions of samples in sample space. Let's start with some motivation. 

You know what they say, "no good deed goes unpunished", and your awesome success with analyzing food data is no different. You now have a bigger task.

```python
# run forest run
import pandas as pd

# generating some random data
n = 5000

# coordinates
x = np.random.normal(0, 0.1, n) + np.random.randint(-1, 1, n)
y = np.random.normal(0, 0.13, n) + np.random.randint(-1, 1, n)

# food features
tasty = np.random.randint(0, 5, n) * 100
healthy = np.random.poisson(1, n) * 100

looking_for_food = pd.DataFrame({'x': x, 'y':y, 'tasty': tasty, 'healthy': healthy})
looking_for_food.head()
```

Try to present the data with a scatter plot the same way we did before

```python
plt.figure()
plt.scatter(looking_for_food.x, looking_for_food.y)

plt.show()
```

That's not very informative, is it? It just looks like a set of blobs...

Here is exactly where histograms come into play. Histograms assignes each data point into a bin and presents us with size of each bin. 

### exercise
1. use `plt.hist` to plot the x values with 2, 13 and 100 bins
2. do the same but for y+x

**Note** use subplots and a loop to keep the plots and code tight

```python
f, ax = plt.subplots(6, 1)

bins = [2,13,100]

for i, bin_size in zip(range(3),bins):
    ax[i].hist(looking_for_food.x, bins = bin_size)

for i, bin_size in zip(range(3),bins):
    ax[i+3].hist(looking_for_food.x+looking_for_food.y, bins = bin_size)
```

I hope we are in agreement that histograms are great. And the greatness doesn't stop here. `plt.hist` actually returns the number of elements per bin and the bin edges as return parameters.

## Adding some words
while a picture is worth a thousand words you still may want to add some words to your plots. You can achieve this using the `plt.text` command. 

### exercise
plot one of the histogtrams again and add some text to highest bin in it

```python
plt.figure()
plt.hist(looking_for_food.x+looking_for_food.y, bins=100)
plt.text(-1.5,200, 'Im on top of the graph and highest bin')
plt.show()
```

Sometimes we want to show to which point exactly our text refers to. To achieve this we can use the `plt.annotate` function. This function puts some text with an arrow pointing to some point in the plot.

### exercise
repeat the previous exercise but now with annotation. You'll need to set the following parameters
1. `xy` for where to point
2. `xytext` for where to put the text
3. `s` the text
4. `arrowprops` a dictionary with arrow properties (e.g., `arrowprops={'arrowstyle':"->"}`)

```python
plt.figure()
plt.hist(looking_for_food.x+looking_for_food.y, bins=100)
plt.annotate(xy=(-1,176), xytext=(-1.5,200), s ='top of the largest bin', arrowprops={'arrowstyle':"->"} )
plt.show()
```

## 2D histograms
While 1D histograms are extremely useful for our case we need a bigger gun, twice as big actually. Use a 2D histogram with `plt.hist2d` to show the density of `(x,y)`. Use a colorbar and play with number of bins to make the result meaningful.

```python
plt.figure()
plt.hist2d(looking_for_food.x,looking_for_food.y, bins=50)
plt.colorbar()
plt.show()
```

# Contours
matplotlib has many many plot types. You probably won't encounter all of them during your studies and work. Just to give a taste of something different, let's do some topography and learn how to plot functions of two parameters.

```python
# creating the two parameters
x = np.linspace(0, 3, 100)
y = np.linspace(-0.5, 1.5, 100)

# creating a 2D grid (matching the shapes of x and y by duplication)
X, Y = np.meshgrid(x, y)
print('X = \n{}\n\nY= \n{}\n'.format(X[:3, :3], Y[:3, :3]))

# conjuring some arbitrary function
Z = np.abs(X-1.5)*np.abs(Y-0.5) + np.sin((X+Y)*np.pi)/2

# plotting
contours = plt.contour(X, Y, Z, cmap='Greens') # with a specifc color map
plt.clabel(contours, inline=True, fontsize=13) # adding some text
plt.colorbar() # showing a color bar
plt.show()
```
