# Know Thy Neighbor - KNN
KNN or K-nearest neighbor is an extremely simple algorithm for supervised learning. KNN answers a question about a data point by using the answers for the k-nearest neighbors for the same question. Let's say you're considering a restaurant for lunch. The KNN approach would be:
1. find the $k$ friends who are most similar to you
2. go to the restaurant which was most recommended by them

### exercise
1. load the dataset from `http://data.princeton.edu/wws509/datasets/salary.dta`
2. take a single row from the table and remove it's rank (feature=rk)
3. use the knn classifier `sklearn.neighbors.KNeighborsClassifier` with 5 neighbors and the **other** rows and try to predict the "missing" rank
    - use `fit` on the rest of the data
    - use `predict` on the selected row

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_stata('http://data.princeton.edu/wws509/datasets/salary.dta')

rank = df.rk
print("Real rank: " + rank[11])
rank = rank.drop(11)
df = pd.get_dummies(df.drop(columns=['rk']))

droped_row = df.iloc[11]
df = df.drop(11)

knn_classifier = KNeighborsClassifier(n_neighbors = 5)

clean_droped_row = pd.get_dummies(droped_row)

knn_classifier.fit(df, rank)

predicted_row = knn_classifier.predict([droped_row])
print("predicted rank: " + predicted_row)
```

# Data generation
at this section we will generate high dimensional random data from two distributions to simulate a high dimensional binary classification problem.

For each class each feature (column) will be sampled from a binomial distribution. A sample binomial distribution with parameters `n, p` corresponds to the number of heads generated by `n` coin tosses of a coin with probability `p` of landing on head. Run the next cell to get a feel of the binomial distribution. Feel free to change the parameters `n` and `p` and see how results change.

```python
import matplotlib.pyplot as plt
# run forest run
m = 1000  # number of samples
n = 100  # number of tosses for each sample
p = 0.2  # probability of each coin landing on head
sample_vec = np.random.binomial(n, p, m)
plt.figure()
plt.hist(sample_vec)
plt.xlim([0, n])
plt.title('Binomial distribution: {} samples, {} tosses, coin parameter (bias) {}'.format(m, n, p))
plt.show()
```

Now for random data generation

```python
# run forest run
m1 = 300  # number of samples (rows) of class 1
m2 = 400  # number of samples (rows) of class 2
n = 1  # number of tosses for all features and classes
p1_vec = np.arange(0.35, 0.66, 0.03)  # coin parameters for class 1
p2_vec = np.arange(0.66, 0.34, -0.03)  # coin parameters for class 2

# data generation for class 1
class1_data = np.zeros((m1, len(p1_vec)))
for ip1, p1 in enumerate(p1_vec):
    class1_data[:, ip1] = np.random.binomial(n, p1, m1)
    
# data generation for class 2
class2_data = np.zeros((m2, len(p2_vec)))
for ip2, p2 in enumerate(p2_vec):
    class2_data[:, ip2] = np.random.binomial(n, p2, m2)
```

We've generated different data for each class. To convince ourselves that it is indeed different let's take a look

```python
# run forest run

# add class indicator to each class
class1_data_with_label = np.hstack([class1_data, np.zeros((m1, 1))])
class2_data_with_label = np.hstack([class2_data, np.ones((m2, 1))])

# create a joined data frame
columns = ['X{}'.format(i) for i in range(len(p1_vec))] + ['class_']  # column labels
ddf = pd.DataFrame(data=np.vstack([class1_data_with_label, class2_data_with_label]), columns=columns)

# plot the data
plt.figure()
plt.pcolormesh(ddf.loc[:, columns[:-1]])
plt.xticks(np.arange(0.5, ddf.shape[1]-1, 1), columns[:-1])
plt.yticks([m1/2, m1 + m2/2], ['class1', 'class2'])
plt.show()
```

# Dimensionality Reduction 
multidimensional data is hard to make sense of and if the dimension is very high, may pose a challenge for many algorithms. There are many methods to reduce the dimensionality of the data. One such method which is very popular for visualization is `t-SNE`. `t-SNE` is a nonlinear dimensionality reduction method which transforms multidimensional data to two or three dimensions such that distances in the new representation fit those in the original. 


## intuition
Ideally a `t-SNE` transform would find a representation in a lower dimension such for each pair of points $ x_1, x_2 $ in the original data the corresponding points in the lower dimension $ y_1, y_2 $ would satisfy $ \lVert x_1 - x_2 \rVert \propto \lVert y_1 - y_2 \rVert $. Or in words a transformation to a lower dimension such that all distances maintain proportion. Unfortunately that is impossible in all but the most trivial of cases. The glory of `t-SNE` algorithm is the clever way it optimizes distances in the lower dimension to **approximate** the ideal.
