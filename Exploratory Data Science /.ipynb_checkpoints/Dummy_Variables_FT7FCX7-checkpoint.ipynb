Some data contains categorical variables with no obvious numeric meaning e.g., city, profession. To allow analysis the most common approach is replacing these features with indicator/dummy variables using `pd.get_dummies(df)`

We will be using a free dataset found in the following address http://data.princeton.edu/wws509/datasets/#salary. Read the data using `pd.read_stata` with the following url `http://data.princeton.edu/wws509/datasets/salary.dta`.

column description:
- sx = Sex, coded 1 for female and 0 for male
- rk = Rank, coded
    - 1 for assistant professor,
    - 2 for associate professor, and
    - 3 for full professor
- yr = Number of years in current rank
- dg = Highest degree, coded 1 if doctorate, 0 if masters
- yd = Number of years since highest degree was earned
- sl = Academic year salary, in dollars.

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_stata('http://data.princeton.edu/wws509/datasets/salary.dta')


df = pd.get_dummies(df)
print(df)
df = pd.get_dummies(df.columns[6])
print(df)
```
