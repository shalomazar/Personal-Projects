```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
import seaborn as sns

csv_file = 'tmdb_5000_movies.csv'
csv_pd = pd.read_csv(csv_file, usecols=['release_date', 'runtime'])

decs = []

csv_pd = csv_pd.loc[csv_pd['runtime'] > 0].loc[csv_pd['runtime'] < 999]
csv_pd = csv_pd.dropna()
csv_pd['runtime'] = csv_pd['runtime']/60

for row in csv_pd.iterrows():
    decade = int(int(row[1][0].split('-')[0])/10) * 10
    decs.append(decade)

csv_pd['decs'] = decs
csv_pd = csv_pd.drop(['release_date'], axis = 1)

csv_pd.boxplot(by='decs')
group_by_df = csv_pd.groupby(['decs'])

sums = group_by_df['runtime'].sum()
print(sums)
```
