# Question 1

Sample from the exponential distribution using numpy.random.exponential (note that the scale parameter is 1/Î»). Plot a histogram of your samples, and calculate the sample mean and variance.

```python
import numpy as np
import matplotlib.pyplot as plt
lamda = 1/40
samples = np.random.exponential(1/lamda, size=100)
plt.hist(samples, bins=5)
plt.show()
sample_mean = samples.mean()
sample_variance = samples.var()
print("sample mean: " + str(sample_mean))
print("sample variance: " + str(sample_variance))
```

# Question 2
1. Derive an analytical solution for the MLE of $\lambda$
2. Plot the error of the MLE as a function of the sample size.

```python
lamdahat = 1/sample_mean
mle_lamda = lamdahat - lamda
print(mle_lamda)
errors = []
for i in range(10,500):
    lamda = 1/40
    samples = np.random.exponential(1/lamda, size=i)
    lamdahat = 1/(samples.mean())
    error = lamdahat - lamda
    errors.append(error)
plt.plot(errors)
plt.show()
```
