# MLE and MAP for Normal Distribution

## Question 1

Read the csv file `triple_vertical.csv` into a data frame. The
description of this data file can be found in `triple_vertical.txt`.
Assume the column ‘vertical’ is distributed normally.

```python
import pandas as pd

df = pd.read_csv("triple_vertical.csv")
```

(a) Calculate MLE for $\mu$ and $\sigma^2$

```python
vertical_mean = df.loc[:,'vertical'].mean()
vertical_variance = df.loc[:,'vertical'].var()
print("MLE for μ: " + str(vertical_mean))
print("MLE for σ2: " + str(vertical_variance))
```

(b) Calculate the estimated standard error for $\mu$

```python
import math
sigma = math.sqrt(vertical_variance)
print("estimated standard error for μ: " + str(sigma))
```

(c) Calculate the approximate 95% confidence interval.

```python
print("approximate 95% confidence interval: " + str(vertical_mean + [-1.96 * sigma, 1.96 * sigma]))
```
